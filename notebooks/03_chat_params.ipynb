```python
from ollama import chat
from ollama import ChatResponse
```

```python
def init_message(system = None):
    message = []
    if system:
        system_message = {
            "role": "system",
            "content": system
        }
        message.append(system_message)
    return message


def add_user_message(message, text):
    user_message = {
        "role": "user",
        "content": text
    }
    message.append(user_message)
    return message

def add_assistant_message(message, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    message.append(assistant_message)
    return message

def chat_with_llm(message, temperature = 1.0, num_predict = 1024, stream = False, think = False, seed = None, top_p = None, top_k = None, repeat_penalty = None):
    params = {
        "model": "qwen3.5:9b",
        "messages": message,
        "stream": stream,
        "think": think,
        "options": {
            "temperature": temperature,
            "num_predict": num_predict
        }
    }
    if seed is not None:
        params["options"]['seed'] = seed
    if top_p is not None:
        params["options"]['top_p'] = top_p
    if top_k is not None:
        params["options"]['top_k'] = top_k
    if repeat_penalty is not None:
        params["options"]['repeat_penalty'] = repeat_penalty
    response: ChatResponse = chat(**params)
    return response['message']['content']
```

```python
messages = init_message(system = "You are a movie idea generator.")

add_user_message(messages, "Generate a sentence of moive idea.")

answer = chat_with_llm(messages)
print(answer)
```

```python
# Seed param test 
message = init_message(system = None)

add_user_message(message, "Generate a sentence of moive idea.")

for i in range(5):
    answer = chat_with_llm(message, seed = 42)
    print(answer)
```

```python
# temperature param test
message = init_message(system = None)
add_user_message(message, "Generate a sentence of moive idea.")
for temp in [0.0, 0.5, 1.0, 1.5]:
    result = (chat_with_llm(message, temperature = temp) for _ in range(3))
    print(f"Temperature: {temp}")
    for r in result:
        print(r)
```

```python
# Top p -> nucleus sampling test
message = init_message(system = None)
add_user_message(message, "Generate a sentence of moive idea.")
for top_p in [0.0, 0.5, 1.0]:
    result = (chat_with_llm(message, temperature = 0.5, top_p = top_p) for _ in range(3))
    print(f"Top p: {top_p}")
    for r in result:
        print(r)
```

```python
# Top k -> top-k sampling test
message = init_message(system = None)
add_user_message(message, "Generate a sentence of moive idea.")
for top_k in [0, 1, 10, 50]:
    result = (chat_with_llm(message, temperature = 0.5, top_k = top_k) for _ in range(3))
    print(f"Top k: {top_k}")
    for r in result:
        print(r)
#Top k = 0 means no top-k sampling. In this case, the model will consider all possible next tokens when generating text, without any restriction on the number of tokens to sample from.
```

```python
# Repeat Penalty test
message = init_message(system = None)
add_user_message(message, "Generate a sentence of moive idea.")
for repeat_penalty in [1.0, 1.5, 2.0]:
    result = (chat_with_llm(message, temperature = 0.5, repeat_penalty = repeat_penalty) for _ in range(3))
    print(f"Repeat Penalty: {repeat_penalty}")
    for r in result:
        print(r)
```
