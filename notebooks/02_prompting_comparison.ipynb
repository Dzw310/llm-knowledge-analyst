```python
from ollama import chat
from ollama import ChatResponse
```

```python
# Zero Shot Prompting
response: ChatResponse = chat(model='qwen3.5:9b', messages=[
  {
    'role': 'user',
    'content': 'Classify the sentiment as Positive, Neutral, or Negative.: The outfit ate, no crumbs.',
  },
],
  think= False)
print(response['message']['content'])
```

```python
response: ChatResponse = chat(model='qwen3.5:9b', messages=[
  {
    'role': 'user',
    'content': '''Classify the sentiment as Positive, Neutral, or Negative.
    Example 1:
    Sentence: I love this product. It works perfectly.
    Answer: Positive

    Example 2:
    Sentence: The delivery arrived late and the box was damaged.
    Answer: Negative

    Example 3:
    Sentence: The movie was fine, nothing special.
    Answer: Neutral

    Now classify:
    Sentence: The outfit ate, no crumbs.
    Answer:''',
  },
],
  think = False)
print(response['message']['content'])
```

```python
# With good shots
response: ChatResponse = chat(model='qwen3.5:9b', messages=[
  {
    'role': 'user',
    'content': '''
    Classify the sentiment as Positive, Neutral, or Negative.
    Example 1:
    Love waiting 40 minutes for cold fries.
    Answer: Negative
    Why: Sarcastic “love.”

    Example 2:
    Sentence: She really did her thing on that project.
    Answer: Positive
    Why: Slang phrase for doing well

    Example 3:
    Sentence: This game is stupidly good.
    Answer: Positive
    Why: “Stupidly” sounds negative, but intensifies praise.

    Now classify:
    Sentence: The outfit ate, no crumbs.
    Answer:
    ''',
  },
],
  think = False)
print(response['message']['content'])
```

```python
# With thinking process
response: ChatResponse = chat(model='qwen3.5:9b', messages=[
  {
    'role': 'user',
    'content': '''Classify the sentiment as Positive, Neutral, or Negative.
    Example 1:
    Sentence: I love this product. It works perfectly.
    Answer: Positive

    Example 2:
    Sentence: The delivery arrived late and the box was damaged.
    Answer: Negative

    Example 3:
    Sentence: The movie was fine, nothing special.
    Answer: Neutral

    Now classify:
    Sentence: The outfit ate, no crumbs.
    Answer:
    Please think step by step.
    ''',
  },
],
  think = False)
print(response['message'].get('thinking', ''))                                                                                                                                                          
print(response['message']['content'])
```
