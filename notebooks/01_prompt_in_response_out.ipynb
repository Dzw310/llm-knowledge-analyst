```python
def add_user_messages(messages, text):
    user_message = {
        "role":"user",
        "content":text
    }
    messages.append(user_message)

def add_assistant_messages(messages, text):
    assistant_message = {
        "role":"assistant",
        "content":text
    }
    messages.append(assistant_message)
def chat_with_llm(messages):
    response: ChatResponse = chat(model='qwen3.5:9b', messages=messages, think = False)
    return response['message']['content']
    
```

```python
messages = []
while True:
    user_input = input(">")
    if user_input == "/quit":
        break
    print(f"> {user_input}")
    print("--------------------------------")
    add_user_messages(messages, user_input)
    response = chat_with_llm(messages)
    add_assistant_messages(messages, response)
    print(response)
    print("--------------------------------")
```
